```python

def visualize_evaluation(result):

    labels = ['Fit', 'Pre', 'gen', 'simp']

    x = np.arange(len(labels))  # the label locations
    width = 0.1  # the width of the bars
    fig, ax = plt.subplots(figsize=(15,10))
#     plt.figure(figsize = ((200,200)))
    rects1 = ax.bar(x - 3/2 * width, result[0], width, label='1st', color='grey')
    rects2 = ax.bar(x - 1/2 * width,
                    result[1], width, label='2nd', color='lightgrey')
    rects3 = ax.bar(x + 1/2 * width,
                    result[2], width, label='3rd', color='darkgrey')
    rects4 = ax.bar(x + 3/2 * width,  result[3], width, label='weighted_avg', color='cyan')

    rects5 = ax.bar(x + 7/2 * width,
                    result[3], width, label='original', color='black')

    # Add some text for labels, title and custom x-axis tick labels, etc.
    ax.set_ylabel('quality')
    ax.set_title('quality measures of clustering ')
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.legend()
    fig.tight_layout()
    plt.show()
```
